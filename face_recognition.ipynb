```python
#face detection by rectangle 
import cv2
fd = cv2.CascadeClassifier(cv2.data.haarcascades +"haarcascade_frontalface_default.xml" )
sd = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_smile.xml")
ed = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_eye.xml")
counter = 0
captured = True
vid = cv2.VideoCapture(0)

while captured:
    flag ,img = vid.read()
    im_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
   # th, img = cv2.threshold(img,78,255,cv2.THRESH_BINARY)

    if flag:
        faces = fd.detectMultiScale(img,1.1,5)
        for x,y,w,h in faces:
            cv2.rectangle(img, pt1=(x,y), pt2=(x+w,y+h), color=(150,0,0),thickness=3)
            cropped_image = img[y:y+h, x:x+w]
            smiles = sd.detectMultiScale(cropped_image,1.1,50)
            eye = ed.detectMultiScale(cropped_image,1.1,50)
            
            if len(faces) == 1:
                    counter += 1
                    print(counter)
                 

                    if counter == 10:
                        captured=False
                    else:
                        counter=0

            for xs,ys,ws,hs in smiles:
                cv2.rectangle(img, pt1=(x+xs,y+ys), pt2=(x+xs+ws,y+ys+hs), color=(0,0,255),thickness=3)

                if len(smiles) == 1:
                     counter += 1
                     print(counter)
                 

                     if counter == 10:
                         captured=False
                     else:
                        counter=0  

            for xe,ye,we,he in eye:
                cv2.rectangle(img, pt1=(x+xe,y+ye), pt2=(x+xe+we,y+ye+he), color=(0,255,0),thickness=3)

                
                
                

                        
                if len(eye) == 1:
                        counter += 1
                        print(counter)
                 

                if counter == 10:
                         captured=False
                else:
                 counter=0     

        cv2.imshow("webcam_image",img)
        # cv2.imshow("smiles",cropped_image)
        key=cv2.waitKey(1)
        if key == ord("j"):
            break
    else:
        break


cv2.destroyAllWindows()
vid.release()

```
